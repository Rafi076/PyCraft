```python
# S M Mohiuddin Rafi
# Recipe Manager
class Recipe:
    def __init__(self, name, ingredients):
        self.name = name
        self.ingredients = ingredients

    def __str__(self):
        return f"{self.name}"

    def get_details(self):
        return f"Recipe: {self.name}\nIngredients:\n{self.ingredients}"


class RecipeManager:
    def __init__(self):
        self.recipes = []

    def add_recipe(self, name, ingredients):
        new_recipe = Recipe(name, ingredients)
        self.recipes.append(new_recipe)
        print(f"\n'{name}' has been added successfully!")

    def remove_recipe(self, name):
        recipe_to_remove = None
        for recipe in self.recipes:
            if recipe.name.lower() == name.lower():
                recipe_to_remove = recipe
                break

        if recipe_to_remove:
            self.recipes.remove(recipe_to_remove)
            print(f"\n'{name}' has been removed.")
        else:
            print(f"\nRecipe '{name}' not found.")

    def clear_recipes(self):
        self.recipes.clear()
        print("\nAll recipes have been cleared!")

    def display_recipes(self):
        if self.recipes:
            print("\nCurrent Recipes:")
            for recipe in self.recipes:
                print(f"- {recipe}")
        else:
            print("\nNo recipes available.")

    def show_recipe_details(self, name):
        found_recipe = None
        for recipe in self.recipes:
            if recipe.name.lower() == name.lower():
                found_recipe = recipe
                break

        if found_recipe:
            print(found_recipe.get_details())
        else:
            print(f"\nRecipe '{name}' not found.")


def main():
    recipe_manager = RecipeManager()

    while True:
        print("\nRecipe Manager Menu:")
        print("1. Add Recipe")
        print("2. Remove Recipe")
        print("3. Clear All Recipes")
        print("4. View All Recipes")
        print("5. View Recipe Details")
        print("6. Exit")

        choice = input("\nChoose an option (1-6): ")

        if choice == "1":
            name = input("Enter the recipe name: ")
            ingredients = input("Enter the ingredients (comma-separated): ")
            recipe_manager.add_recipe(name, ingredients)
        elif choice == "2":
            name = input("Enter the recipe name to remove: ")
            recipe_manager.remove_recipe(name)
        elif choice == "3":
            recipe_manager.clear_recipes()
        elif choice == "4":
            recipe_manager.display_recipes()
        elif choice == "5":
            name = input("Enter the recipe name to view details: ")
            recipe_manager.show_recipe_details(name)
        elif choice == "6":
            print("\nExiting the Recipe Manager...")
            break
        else:
            print("\nInvalid choice! Please choose a valid option.")

if __name__ == "__main__":
    main()
```
